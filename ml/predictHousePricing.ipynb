```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Example data: sizes of homes (in square feet) and their corresponding prices (in $1000s)
home_sizes = np.array([750, 800, 850, 900, 950, 1000, 1050, 1100, 1150, 1200]).reshape(
    -1, 1
)
home_prices = np.array([150, 160, 165, 180, 185, 190, 200, 210, 220, 230])

X_train, X_test, y_train, y_test = train_test_split(
    home_sizes, home_prices, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_train)

print(f"Model Cofficient (slope): {model.coef_[0]}")
print(f"Model Intercept: {model.intercept_}")

predicted_price = model.predict([[1150]])[0]
print(f"Predicted price for a 1150 sp ft home: ${predicted_price * 1000:.2f}")
```

```python
# Visulasation
plt.scatter(home_sizes, home_prices, color="blue", label="Actual prices")
plt.plot(home_sizes, model.predict(home_sizes), color="red", label="Regression line")
plt.xlabel("Size of Home(sq ft)")
plt.ylabel("price($1000)")
plt.title("Home Size vs Price")
plt.legend()
plt.show()
```
